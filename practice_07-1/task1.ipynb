- ### *Создайте вектор с помощью библиотеки np (векторы со значением вариант №…),*

Импортируем библиотеку **numpy**

```python
import numpy as np
```

Создадим вектор с данными

```python
vector = np.array([4, 6, 3, 23, 56, 1, 9, 8])
```

```python
vector
```

- ### *Проиндексировать вектор по элементам (от 3х до 5 значений)*

Вектор можно индексировать, используя квадратные скобки с положительными или отрицательными индексами аналогично списку. Проверим это

```python
print(vector[0])
print(vector[2])
print(vector[5])
print(vector[-2])
print(vector[len(vector) - 1])
```

- ### *Проиндексировать по диапазону*

Индексировать также можно и по диапазону, используя срезы. Проверим это на нескольких разных вариантах

```python
print(vector[:4])
print(vector[5:])
print(vector[3:6:2])
print(vector[:3:1])
```

- ### *Перевернуть массив*

Перевернуть массив (вектор) можно, применив функцию flip, посмотрим как она работает

```python
vector = np.flip(vector)
```

```python
vector
```

- ### *Произвести операцию присвоения*

Значения вектора можно переприсваивать любыми другими значениями. Проверим это на нескольких из них

```python
vector[0] = 1
vector[len(vector) - 1] = len(vector)
```

```python
vector
```

- ### *Выполнить математическую операцию деления*

С вектором можно производить большинство математических операций, также как и с числами. При делении вектора все его элементы делятся на делимое

```python
vector = vector / 4
```

```python
vector
```

- ### *Применить методы argmax и argmin*

argmax и argmin работают просто - они возвращают индексы максимального и минимального элементов соответственно

```python
amax = vector.argmax()
amin = vector.argmin()
print(amax)
print(amin)
```

- ### *Применить метод append*

Функция append аналогична одноименному методу списка: он добавляет в конец структуры данных элемент. Она принимает два аргумента: массив в который нужно поместить новое значение и самое это значение.

```python
vector = np.append(vector, values=60)
```

```python
vector
```

В качестве параметра можно передать не только одно значение, но и список

```python
vector = np.append(vector, values=[9, 43, 5])
```

```python
vector
```
