Импортируем библиотеки **pandas, matplotlib**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
extreme_pollution_file_path = (
    "./data/Zagryaznenie_poverkhnostnih_vod_RF_176_23.09.21/extreme_pollution.csv"
)
df = pd.read_csv(extreme_pollution_file_path, sep=";", encoding="utf-8")
```

Обработаем данные

```python
df.info()
```

Преобразуем некоторые столбцы в **категориальный тип**

```python
df["subject"] = df["subject"].astype("category")
df["indicator"] = df["indicator"].astype("category")
df["hazard_class"] = df["hazard_class"].astype("category")
df["unit"] = df["unit"].astype("category")
```

```python
df.info()
```

- ### *Cоздать столбчатую диаграмму и разместить в горизонтальном положении (какие значения будете выносить на график определите сами)*

Создадим диаграмму по количеству случаев загрязнения в регионах

```python
df_barh = df.groupby("subject")["cnt_cases"].sum().sort_values(ascending=False)

df_barh = df_barh.head(10)

plt.figure(figsize=(10, 8))
plt.barh(df_barh.index, df_barh.values)

plt.xlabel("Количество случаев")
plt.title("Количество случаев загрязнения по субъектам РФ")

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

- ### *Создайте два линейных графика одновременно для сравнения исходя из своего варианта (значения для графиков выберите на свое усмотрение)*

Построим линейный график для двух индикаторов (Марганец и Железо общее). По оси Y — общее количество случаев (cnt_cases) за год, по оси X — годы.

Для начал необходимо подготовить данные. Нужно преобразовать колонку period в дату

```python
df["period"] = pd.to_datetime(df["period"])
```

```python
# выделяем год
df["year"] = df["period"].dt.year

indicators = ["Марганец", "Железо общее"]
df_sel = df[df["indicator"].isin(indicators)]

columns_group = ["year", "indicator"]
df_grouped = df_sel.groupby(columns_group)["cnt_cases"].sum().reset_index()

df_mn = df_grouped[df_grouped["indicator"] == "Марганец"]
df_fe = df_grouped[df_grouped["indicator"] == "Железо общее"]

plt.plot(df_mn["year"], df_mn["cnt_cases"], marker="o", label="Марганец")
plt.plot(df_fe["year"], df_fe["cnt_cases"], marker="o", label="Железо общее")

plt.xlabel("Год")
plt.ylabel("Суммарное количество случаев загрязнений")
plt.title("Годовое количество случаев по индикаторам")
plt.legend()
plt.grid(True)

plt.show()
```

- ### *Создать диаграмму рассеивания*

Построим диаграмму рассеивания (Scatter plot), которая показывает связь между двумя количественными переменными (value_min и value_max)
