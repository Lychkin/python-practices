- ### *Проверьте количество сессий*

Создадим функцию count_sessions, которая суммирует количество сессий во всех частях файла и использует рекурсию для обхода всех вложенных элементов

```python
def count_sessions(obj):
    count = 0
    if isinstance(obj, dict):
        if "sessions" in obj and isinstance(obj["sessions"], list):
            count += len(obj["sessions"])
        if "session" in obj:
            count += 1
        for v in obj.values():
            count += count_sessions(v)
    elif isinstance(obj, list):
        for item in obj:
            count += count_sessions(item)
    return count
```

Переберём через цикл каждый файл и применим для него функцию для подсчёта сессий count_session

```python
for path in FILE_PATHS:
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
        print(f"Файл: {path}. Число сессий: {count_sessions(data)}")
```
