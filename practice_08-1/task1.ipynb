- ### *Считайте строки файла методом readlines, опишите что получилось*

Импортируем все необходимые зависимости

```python
import re
from collections import Counter
import pandas as pd
import nltk

nltk.download("punkt_tab")
nltk.download("averaged_perceptron_tagger")
nltk.download("wordnet")
nltk.download("stopwords")
```

С помощью контекстного менеджера открываем файл на чтение и сохраняем результат в переменную. Убедимся что текст из файл считался верно.

```python
FILE_PATH = "data/javascript-введение.txt"

with open(FILE_PATH, "r", encoding="utf-8") as file:
    lines = file.readlines()

lines
```

- ### *Посчитайте самые популярные слова и их частоту в вашем тексте*

Объединим все строки файла для удобства дальнейшней работы


```python
text = " ".join(lines)

text
```

Заменяе всю пунктуацию и спецсимволы на пробел

```python
text = re.sub(r"[^\w\s]", " ", text)

text
```

Токениризуем все слова текста

```python
tokenize = nltk.tokenize.word_tokenize

tokens = tokenize(text.lower(), language="russian")

tokens
```

Используем класс Counter для подсчета слов (токенов)

```python
counter = Counter(tokens)

counter
```

- ### *Выведите в формате DataFrame 10 самых популярных слов*

Создадим датафрейм с частотой слов и посмотрим как он выглядит

```python
df_words = pd.DataFrame.from_dict(dict(counter), orient="index", columns=["count"])

df_words.head()
```

Отсортируем датафрейм по частоте по убывания и выведем 10 первых строк

```python
df_words.sort_values(by="count", inplace=True, ascending=False)

df_words.head(10)
```

- ### *Сохраните полученный DataFrame в формате .csv*

Сохраним результат методом to_csv

```python
df_words.to_csv("data/words.csv")
```

- ### *Сохраните в формате .xlsx.*

Сохраним результат методом to_excel

```python
df_words.to_excel("data/words.xlsx")
```
